```python
search_contig= "Geo_Pool_F_GGCTAC_L006_R1_001_val_1_(paired)_contig_105"
```

```python
for seq_record in SeqIO.parse(datafile, "fasta"):
    if seq_record.id== search_contig:
        print("> "+seq_record.id+ "\t"+str(len(seq_record)))
        print(seq_record.seq)
```

```python
sec=seq_record.seq
sec_length=len(sec)
count_c=sec.count("C")
count_g=sec.count("G")
count_cg=float(sec.count("CG"))
cpgoe=((count_cg/(count_c*count_g))*(sec_length**2/(sec_length-1)))
print(sec)
print(count_c)
print(count_g)
print(count_cg)
print((count_cg/(count_c*count_g)))
print(sec_length**2/sec_length-1)
print(cpgoe)
```
